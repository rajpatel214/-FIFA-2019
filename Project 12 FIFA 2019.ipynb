```python
# defining a function for cleaning the wage column

def remove_symbol_from(value):
    out=value.replace("€","")
    if "M" in out:
        out=float(out.replace("M",""))*1000000
    elif "K" in out:
        out=float(out.replace("K",""))*1000
    return float(out)
    
ff["Value"]=ff["Value"].apply(lambda x: remove_symbol_from(x))
```
